```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('numerical_df.csv')
```

```python
for column in data.columns:
    sns.histplot(data=data[column], kde=True)
    plt.show()
# some like exponential distribution
```

```python
#log transform
transformed_data = np.log(data)
transformed_data
```

```python
inf_vals = transformed_data.columns[transformed_data.isin([np.inf, -np.inf]).any()]
inf_vals
```

```python
transformed_data.replace([np.inf, -np.inf], 0, inplace=True)
transformed_data
```

```python
for column in transformed_data.columns:
    sns.histplot(data=transformed_data[column], kde=True)
    plt.show()
    
#possible exponential distribution in months since policy inception 
```

```python
#square root transformation
transformed_data1 = data.applymap(np.sqrt)
```

```python
for column in transformed_data1.columns:
    sns.histplot(data=transformed_data1[column], kde=True)
    plt.show()
```

```python
#quantile transformation
from sklearn.preprocessing import QuantileTransformer
quantile_transformer = QuantileTransformer(output_distribution='normal')
transformed_data3 = quantile_transformer.fit_transform(data)
transformed_data3 = pd.DataFrame(transformed_data3, columns=data.columns)
```

```python
for column in transformed_data3.columns:
    sns.histplot(data=transformed_data3[column], kde=True)
    plt.show()

# some like! normal distributions
```
